```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient("mongo")
database = client['exercices']
collection = database['youtube']
```

```python
collection.drop()
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(2)
```

## Question 0

```python
df_json=pd.read_json("./data/US_category_id.json")
df_json["category_id"]=df_json["items"].apply(lambda x: int(x["id"]))
df_json.head()
```

### Importer les données

```python
df_youtube.merge(df_json, on='category_id').head()
```

```python
df_youtube["tags"] = df_youtube["tags"].apply(lambda x: x.split("|"))
```

```python
df_youtube["tags"]
```

```python
collection.insert_many(df_youtube.to_dict(orient='records'))
```

```python
col=collection.find()
next(col)
```

## Question 4

```python
views=collection.find().sort("views",-1).limit(5)
list(views)
```

## Question 5

```python
av_views=collection.aggregate([{"$group":{"_id":"$category_id", "average_views":{"$avg":"$views"}}}])
list(av_views)
```
